# Analise e apresentação basica dos dados

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.data = pd.to_datetime(df.data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df.data, df.temperatura, color = 'crimson', lw = 1,  ls = 'dotted')
#documentação com as cores https://matplotlib.org/stable/gallery/color/named_colors.html
#lw = 1,  ls = 'dotted', marker = 'o' espessurada linha, tipo da linha e marcadores
eixo.set_title('Temperatura no tempo', fontsize = 25)
eixo.set_xlabel('Data',fontsize = 20)
eixo.set_ylabel('Temperatura',fontsize = 20)
eixo.legend(['temperatura'], loc = 'lower right', fontsize = 15)
```

# Mais de um eixo na mesma figura

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color = 'crimson')
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.set_title('Temperatura em 2014', fontsize=25, pad=20)
eixo.legend(['temperatura'], loc='lower right', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)

eixo2.plot(df['data'], df['temperatura'], color = 'purple')
eixo2.set_title('Temperatura no momento', fontsize=15)
eixo2.legend(['temperatura'], loc='best', fontsize=8)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.grid(True)
```

# Analise do mes de Maio de 2014

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

#Eixo 1
eixo.plot(df.data, df.temperatura, color = 'crimson')
eixo.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,5,31))
eixo.set_ylim(270, 320)
eixo.set_title('Temperatura em maio/2014', fontsize=25, pad=20)
eixo.legend(['temperatura'], loc='lower right', fontsize=15)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.grid(True)

#filtro
purple_esquerda = df.data < datetime.datetime(2014,5,1)
purple_direita = df.data > datetime.datetime(2014,5,31)

#Eixo 2
eixo2.plot(df.data, df.temperatura, color = 'crimson')
eixo2.plot(df[purple_esquerda]['data'], df[purple_esquerda]['temperatura'], color = 'purple')
eixo2.plot(df[purple_direita]['data'], df[purple_direita]['temperatura'], color = 'purple')
eixo2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2014,12,31))
eixo2.set_title('Temperatura em 2014', fontsize=15)
eixo2.legend(['temperatura'], loc='best', fontsize=8)
eixo2.set_ylabel('Temperatura', fontsize=10)
eixo2.set_xlabel('Data', fontsize=10)
eixo2.grid(True)
```

# Maximas e minimas de maio de 2014

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'crimson')
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

#seta indicando o ponto de máximo
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) -5
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

#seta indicando o ponto de máximo
x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura']) + 5
eixo.annotate('Mínimo', xy=(x1,y1), fontsize=20, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

#linha de maximo e minimo
eixo.axhline(max(df['temperatura']), color = 'k', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'k', linestyle='--')

```

# Análise exploratória das temperaturas nos dias da semana

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
cores = ['crimson','r', 'orange', 'yellow', 'green', 'b', 'purple']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor = 'k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
explodir = [0.1,0,0,0,0,0.1,0.1]
eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index, autopct = '%.1f%%',explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', size=15, pad=10)
```
